# Webscrapping

web scraping is another method to get data from internet

```python
webpagerequest = requests.get("https://www.rki.de/DE/Content/InfAZ/N/Neuartiges_Coronavirus/Fallzahlen.html")
```

```python
webpagerequest
```

```python
Soupcreat = BeautifulSoup(webpagerequest.content, 'html.parser') # beautifulsoap is used to pass every thing in elegant way.
```

```python
Soupcreat.get_text()
```

```python
webpage1_table=Soupcreat.find('table')
```

```python
all_rows=webpage1_table.find_all('tr')
Final_list =[]
```

```python
for pos,rows in enumerate(all_rows):
    #print(pos)
    #print(rows)
    col_list=[each_col.get_text(strip=True) for each_col in rows.find_all('td')] 
    #print(col_list)
    Final_list.append(col_list)
    
```

```python
Final_list
```

```python
#add data to data frame and rename columns 
#dropna() to remove rows
updated_data_status=pd.DataFrame(Final_list).dropna().rename(columns={0:'State',
                                                       1:'cases',
                                                       2:'changes in cases per day',
                                                       3:'cases_per_100k',
                                                       4:'Death',
                                                       5:'comments'})
```

```python
updated_data_status
```

```python
updated_data_status.head()
```
